### 01. Customizing plots

### Line Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,201)
y = np.sin((2*np.pi*x)**2)
plt.plot(x, y, 'red')
plt.show()
```

### Scatter plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 10 * np.random.rand(200,1)
y = (0.2 + 0.8 * x) * np.sin(2*np.pi*x) \
    + np.random.randn(200,1)
```

```python
plt.scatter(x,y)
plt.show()
```

### Histograms 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 10 * np.random.rand(200,1)
y = (0.2 + 0.8 * x) * \
     np.sin(2*np.pi*x) + \
     np.random.randn(200,1)
```

```python
plt.hist(y, bins=20)
plt.show()
```
