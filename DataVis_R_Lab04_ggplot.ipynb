### ggplot 

```r
library(ggplot2)
library(MASS)
data(Cars93)
```

### barplot

```r
Plot <- ggplot( Cars93, aes(x=Type, y=Horsepower, fill=DriveTrain)) 
Plot + geom_bar(position="dodge", stat="identity")
```

```r
### 
Plot + geom_bar(stat="identity", position = "stack") # stack 쌓기, dodge 옆으로
```

```r
Plot + geom_bar(stat="identity", position = "dodge") # stack 쌓기, dodge 옆으로
```

### fill 확률 표시

```r
Plot + geom_bar(stat="identity", position = "fill") # stack 쌓기, dodge 옆으로, 
```

### 가로형 막대 그래프

```r
Plot + geom_bar(stat="identity", position = "fill") + coord_flip()
```
