## 02. 군집분석(Clustering)

```r
library(igraph)
df <- read.csv("D:/dataset/RLoveYou/clustering.csv")
head(df,10)
graph <- data.frame(학생=df$학생, 교수=df$교수)
```

```r
library(stringr)
df <- graph.data.frame(graph,directed=TRUE)  
df
 
gubun1 <- V(df)$name
head(gubun1,100)          # 이름 출력
gubun <- str_sub(gubun1, start=1, end=1)
table(gubun)
```

```r
plot(df)
```

```r
plot(df, layout=layout.fruchterman.reingold,
     vertex.size=2,            # 점의 크기
     edge.arrow.size=0.2,      # 화살표 크기
     vertex.color = "green",   # 점의 색깔
     vertex.label=NA)
```

### 학생과 교수의 색상과 크기를 구분해서 출력

```r
colors <- c()
for (i in 1:length(gubun)) {
    if (gubun[i]=='S') {
        colors <- c(colors, "red") 
    } else {
        colors <- c(colors, "green")
    }
}
```

```r
sizes <- c()
for (i in 1:length(gubun)) {
    if (gubun[i]=='S') {
        sizes <- c(sizes,2)
    } else {
        sizes <- c(sizes,6)
    }
}
```

```r
plot(df, layout=layout.fruchterman.reingold,
     vertex.size=sizes,            # 점의 크기
     edge.arrow.size=0.2,      # 화살표 크기
     vertex.color = colors,   # 점의 색깔
     vertex.label=NA)
```
